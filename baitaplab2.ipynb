```python
def Q1(g):
  return np.percentile(g, 25)
```

```python
def Q2(g):
  return np.percentile(g, 50)
```

```python
def Q3(g):
  return np.percentile(g, 75)
```

```python
def std(g):
  return np.std(g)
```

```python
# Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1,
# Q2 và Q3 của DH1 theo KT
pivot = df.pivot_table(values= "DH1", index=["KT"], aggfunc=["count", "sum", "mean", "min","max",std,Q1,Q2,Q3])
pivot
```

```python
# Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1, Q2, Q3
pivot2 = df.pivot_table(values= "DH1", index=["KT", "KV"], aggfunc=["count", "sum", "mean", "min","max",std,Q1,Q2,Q3])
pivot2
```

```python
# Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1, Q2
pivot3 = df.pivot_table(values= "DH1", index=["KT", "KV", "DT"], aggfunc=["count", "sum", "mean", "min","max",std,Q1,Q2,Q3])
pivot3
```
